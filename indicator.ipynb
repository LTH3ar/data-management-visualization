## Data Cleaning
In the above stats we can see that there are duplicate but that are common for large survey, we are going to drop some column we don't need

```python
# combine Fruits and Veggies
df['Fruits_or_Veggies'] = (
    (df['Fruits'] == 1) | 
    (df['Veggies'] == 1)
).astype(int)

# remove unused column
df_cleaned = df.drop(columns=['CholCheck', 'Stroke', 'HeartDiseaseorAttack', 'DiffWalk', 'MentHlth', 'PhysHlth', 'Fruits', 'Veggies'])
# print(df_cleaned.head().to_csv(index=False))
df_cleaned.head()
```

## Features Engineering

```python
# 1. Create composite scores (similar to your PIMA approach)
df_cleaned['Behavioral_Risk_Score'] = (
    (df_cleaned['PhysActivity'] == 0).astype(int) +  # No exercise
    (df_cleaned['Smoker'] == 1).astype(int) +        # Smoker
    (df_cleaned['HvyAlcoholConsump'] == 1).astype(int) + # Heavy drinker
    (df_cleaned['Fruits_or_Veggies'] == 0).astype(int)
)

# 2. BMI Categories (match PIMA)
df_cleaned['BMI_Category'] = pd.cut(df_cleaned['BMI'], 
                                            bins=[0, 18.5, 25, 30, 100],
                                            labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

# 3. Decode Age groups for interpretability
age_mapping = {1: '18-24', 2: '25-29', 3: '30-34', 4: '35-39', 5: '40-44',
               6: '45-49', 7: '50-54', 8: '55-59', 9: '60-64', 10: '65-69',
               11: '70-74', 12: '75-79', 13: '80+'}
df_cleaned['Age_Group'] = df_cleaned['Age'].map(age_mapping)

# 4. Income levels (for disparity analysis)
income_mapping = {1: '<$10k', 2: '$10k-15k', 3: '$15k-20k', 4: '$20k-25k',
                  5: '$25k-35k', 6: '$35k-50k', 7: '$50k-75k', 8: '$75k+'}
df_cleaned['Income_Level'] = df_cleaned['Income'].map(income_mapping)

# 5. Healthcare access composite
df_cleaned['Healthcare_Barrier'] = (
    (df_cleaned['AnyHealthcare'] == 0).astype(int) +
    (df_cleaned['NoDocbcCost'] == 1).astype(int)
)

df_cleaned.head()
```
